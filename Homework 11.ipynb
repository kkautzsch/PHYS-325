```python
#Kaleb Kautzsch
#Part (i)

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.constants import pi
from numpy import sqrt, exp

x = np.random.randint(0,10,50)

def p(x):
    return (2/sqrt(2*pi))*exp(-x**2/2)

y = np.linspace(0,4,50)

plt.plot(y,p(x))
plt.title("Probability Density")
```
