<b>Exercise 1</b> <i>Continuous Fourier transforms</i>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, pi
from scipy.fftpack import fft, fftfreq
from scipy.fftpack import fftshift, ifftshift

def f(t):
    return (sin(t))**2/t**2


def ftrans(omega):
    return pi*np.maximum(1-abs(omega/2),0)

rangeX = 10
dx=0.001
t = np.arange(-rangeX,rangeX,dx)
N = len(t)

def ftf(omega):
    return np.pi*np.maximum(1-np.abs((omega/2)),0)

def myFFT(rangeX, dx):
    
    dFreq = 2*np.pi/dx/N
    freqMax = dFreq*N/2
    freqGauss = np.arange(-freqMax,freqMax,dFreq)

    fftGauss = fftshift( fft( ifftshift(f(t)) ) )

    ftNumerical = np.real(dx*fftGauss)

    return freqGauss, ftNumerical

numSol = myFFT(N,dx)

plt.plot(t,f(t), label ='original function')
plt.plot(t,ftf(t), label ='analytical solution')
plt.plot(numSol[0],numSol[1], '--', label="numerical solution")

plt.xlim(-rangeX,rangeX)
plt.ylim(0,pi)

plt.legend(loc = "upper right")
```

<b>Exercise 3</b> <i>Data smoothing with Fourier transforms</i>

```python
#Kaleb Kautzsch
#Part (i)

import numpy as np
import matplotlib.pyplot as plt

dow = np.loadtxt("dow.txt",float)

plt.plot(dow,color='red')
plt.title('Dow Jones Industrial Average')
plt.xlabel('Days')
plt.ylabel('Daily Closing Value')
```

```python
#Kaleb Kautzsch
#Part(ii)

from scipy.fftpack import fft, ifft, fftshift, ifftshift

c = fft(dow)
moddingC1 = c
moddingC1[52:972] = 0
dowMod1 = ifft(moddingC1)

plt.plot(dowMod1)
```

```python

c = fft(dow)
cMod = c
cMod[10:1014] = 0
dowMod =ifft(cMod)

plt.plot(dowMod)
```

```python
#Kaleb Kautzsch
#Part (iii)

plt.plot(dow)
plt.plot(dowMod1)
plt.plot(dowMod)

print("The degree of smoothing can be influenced by removing the number of secondary peaks i.e")
print("the more peaks removed, the smoother the function.")
```
