<b>Energy Levels of an Electron</b>

```python
#Import programs
import numpy as np

#Variables
m_e = float(9.11e-31) #Mass of electron in kg
h_bar = float(1.05e-34) #Plancks constant in J*s
pi = float(3.14) #Pi
wellWidth = float(input("Please state the well's width in meters: "))
qNum = int(input("Please state the quantum number: ")) #qNum is the inputed quantum number

#Formula function for energy
numerator = np.square(qNum*h_bar*pi)
denominator = ((2*m_e)*(wellWidth**2))
ans = numerator/denominator

#If, Else statements for the function
if wellWidth < 0:
    print("Your well is too small. Try again. ")
elif qNum <= 0:
    print("Sorry you entered a bad number. Try again. ")
else:
    print("The energy of the well is " + str(ans) + " Joules. ")
```

<b>Farenheit to Celsius Conversion</b>

```python
##Temperature Conversion Formulas
farenheit = int(input("Please state how hot/cold it is: "))
celcius = (farenheit - 32) * 5/9

#Printed Conversion
print("Temperature in Celcius is: " + str("%.2f" % celcius) + " C ")
```

<b>Reading and Plotting Data</b>

```python
##Import programs
import numpy as np
import matplotlib.pyplot as plt
import csv

#Empty arrays for the data to be passed into
time = []
humidity = []
temp = []

#Passing in the file for Python to read and format
with open('weather_data.txt', 'r') as csvfile:
    """The plots variable will change the file type and separate the data"""
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        time.append(float(row[0]))
        humidity.append(float(row[2]))
        temp.append(float(row[1]))

#Making the plot for Temperature/Time
plt.plot(time,temp)
plt.xlabel('Unix Time (s)')
plt.ylabel('Temperature (C)')
plt.title('Temperature vs. Time')
plt.show()

#Making the plot for Humidity/Time
plt.plot(time,humidity)
plt.xlabel('Unix Time (s)')
plt.ylabel('Humidity (%)')
plt.title('Humidity vs. Time')
plt.show()

#Calculating the Averages
print("The average temperature is: " + str(np.mean(temp)) + " C")
print("The average humidity is: " + str(np.mean(humidity)) + "% ")
print("The max temperature is " + str(np.max(temp)) + ". The minimum temperature is " + str(np.min(temp)))
```
