<b>Exercise 1</b> <i>Interpolation</i>

```python
#Kaleb Kautzsch
#Part (i)

#Import libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

x = np.array([-4,-3,-2,-1,0,1,2,3,4])
y = np.array([1.125e-7, 1.234e-4, 1.832e-2, 3.679e-1, 1, 3.679e-1, 1.832e-2, 1.234e-4, 1.125e-7])

pol = np.polyfit(x,y,len(x)-1)

xx = np.linspace(min(x),max(x))
yy = np.polyval(pol,xx)

plt.plot(xx, yy, '-',label='Linear Fit')
plt.axis([min(xx), max(xx), min(yy), max(yy)])

#Part(ii)
xRange = np.arange(-4.0, 4.0, 0.01)
splineCubic = interp1d(x, y, kind='cubic')

plt.plot(x,y, label='Points Given')
plt.plot(xRange, splineCubic(xRange), '--', label='Cubic Fit')
plt.xlabel("x",fontsize=16)
plt.ylabel("y",fontsize=16)
plt.title("8th Degree Polonomial Interpolation", fontsize=18)
leg = plt.legend()
plt.show()
```

<b>Exercise 2</b> <i>Probability and Statistics</i>

```python
#Kaleb Kautzsch
#Part(i)-Part(ii)

#Import libraries
import random
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

outcomes = ('000', '001', '010', '011', '100', '101', '110', '111')
tails = ('0','1','2','3')
y = np.zeros([8])
z = np.zeros([4])

for i in range(10000):
    headTail1 = random.randint(0,1)
    headTail2 = random.randint(0,1)
    headTail3 = random.randint(0,1)
    if headTail1 == 0:
        if headTail2 == 0:
            if headTail3 == 0:
                y[0] += 1
                z[0] += 1
            if headTail3 == 1:
                y[1] += 1
                z[1] += 1
        if headTail2 == 1:
            if headTail3 == 0:
                y[2] += 1
                z[1] += 1
            if headTail3 == 1:
                y[3] += 1
                z[2] += 1
    if headTail1 == 1:
        if headTail2 == 0:
            if headTail3 == 0:
                y[4] += 1
                z[1] += 1
            if headTail3 == 0:
                y[5] += 1
                z[2] += 1
        if headTail2 == 1:
            if headTail3 == 1:
                y[6] += 1
                z[2] += 1
            if headTail3 == 1:
                y[7] += 1
                z[3] += 1

plt.bar(outcomes,y)
plt.xlabel('Outcomes')
plt.ylabel('Occurrences')
plt.title('Flip reults')
```

```python
#Bar plot of how many tails there were per outcome
plt.bar(tails, z)
histogram = np.histogram(tails,bins=[0,1,2,3])
plt.xlabel('Outcomes')
plt.ylabel('Occurrences')
plt.title('Tails per set')
print(histogram)
```
