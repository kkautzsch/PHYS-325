```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, pi
from scipy.fftpack import fft, fftfreq
from scipy.fftpack import fftshift, ifftshift

def f(t):
    return (sin(t))**2/t**2


def ftrans(omega):
    return pi*np.maximum(1-abs(omega/2),0)

rangeX = 10
dx = 0.001
t = np.arange(-rangeX,rangeX,dx)
N = len(t)

def ftf(omega):
    return np.pi*np.maximum(1-np.abs((omega/2)),0)

def myFFT(rangeX, dx):
    
    dFreq = 2*np.pi/dx/N
    freqMax = dFreq*N/2
    freqGauss = np.arange(-freqMax,freqMax,dFreq)

    fftGauss = fftshift( fft( ifftshift(f(t)) ) )

    ftNumerical = np.real(dx*fftGauss)

    return freqGauss, ftNumerical

numSol = myFFT(N,dx)

plt.plot(t,f(t), label =r"$\sinc^2(t)$")
plt.plot(t,ftf(t), label ='analytical solution')
plt.plot(numSol[0],numSol[1], '--', label="numerical solution")
plt.xlim(-rangeX,rangeX)
plt.ylim(0,pi)
plt.legend(loc = "upper right")
plt.show()
```
