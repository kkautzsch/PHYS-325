```python
#Kaleb Kautzsch
import math

#Constants
h_bar = 1.055e-34 #Plancks reduced constant in J*s
m_e = 9.11e-31 #Mass of electron in kg
pi = math.pi

#Input Variables
n = int(input("Input a quantum number: ")) #Quantum number value asked by user
a = float(input("Input a well width in meters and using scientific notation: "))

#Formula
numerator = n**2*h_bar**2*pi**2
denominator = 2*m_e*a**2
E = numerator/denominator

print("The energy from inputs",n,"and",a,"m equals",E, "J") 
```

I am not sure why, but the code needs to ran twice in order for the plots to show.

```python
#Kaleb Kautzsch
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

time = []
temp = []
humid = []
timechange = []

rawdata = open("weather_data.txt",'r')
weatherdata=rawdata.readlines()
data = weatherdata
rawdata.close()
fig, ax1 = plt.subplots()


#Iterating and appending over the data set
for i in range(len(weatherdata)):
    weatherdata=data[i]
    rows=weatherdata.split(',')
    time.append(float(rows[0]))
    timechange.append((time[i]-1503187200)/3600)
    temp.append(float(rows[1]))
    humid.append(float(rows[2]))
    

#Creating the first plot of Temp/Time
ax1.set_xlabel('Time (Hours after August 20, 2017)')
ax1.set_ylabel('Temperature (C)',color='b')
ax1.plot(timechange,temp, color='b', label='Temp')
ax1.tick_params(axis='y', color='b')


#Making the second plot share the same x-axis
ax2 = ax1.twinx()

#Creating the second plot Humidity/Time
ax2.set_ylabel('Humidity (%)', color='r')
ax2.plot(timechange,humid, color='r', label='Humidity')
ax2.tick_params(axis='y', color='r')


fig.tight_layout() #To eliminate any weird cutoff on y-axis'
fig.legend(loc='upper left')
fig.savefig('weatherplots.png')

print("The average temperature is",np.mean(temp),"Centigrade")
print("The highest temperature is",np.max(temp),"Centigrade and the lowest temperature is",np.min(temp),"Centigrade")
print("The average humidity is",np.mean(humid),"%")
```
