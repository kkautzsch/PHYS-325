This polynomial is a degree 8 polynomial.

```python
#@author Cody Fialkowski

import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
from scipy import interpolate

dataPoints = np.array([[-4, 1.125e-7],
                      [-3, 1.234e-4],
                      [-2, 1.832e-2],
                      [-1, 3.679e-1],
                      [0, 1],
                      [1, 3.679e-1],
                      [2, 1.832e-2],
                      [3, 1.234e-4],
                      [4, 1.125e-7]])
x = dataPoints[:,0]
y = dataPoints[:,1]
sigmaY = np.std(y)

pol = np.polyfit(x,y,len(x)-1) 

xx = np.linspace(min(x),max(x))
yy = np.polyval(pol,xx)  

xrange = np.arange(-4.,4.,.01)

splineL = interpolate.interp1d(x,y,kind='linear')
splineC = interpolate.interp1d(x,y,kind='cubic')

plt.plot(x, y, '*m', markersize = 10)
plt.plot(xx, yy, 'g', label="polynomial")
plt.plot(xrange, splineL(xrange), 'r', label='linear')
plt.plot(xrange, splineC(xrange), 'b', label='cubic')
plt.title("i put purple stars for my points be proud of me")
plt.xlabel("this is what axis x is on")
plt.ylabel("this is what axis y is on")
leg = plt.legend()
```

```python
#@author Cody Fialkowski

import random
import numpy as np
import matplotlib.pyplot as plt

outcomes = ('000', '001', '010', '011', '100', '101', '110', '111')
tails = ('0','1','2','3')
y = np.zeros([8])
z = np.zeros([4])

for i in range(10000):
    r = random.randint(0,1)
    s = random.randint(0,1)
    t = random.randint(0,1)
    if r == 0:
        if s == 0:
            if t == 0:
                y[0] += 1
                z[0] += 1
            if t == 1:
                y[1] += 1
                z[1] += 1
        if s == 1:
            if t == 0:
                y[2] += 1
                z[1] += 1
            if t == 1:
                y[3] += 1
                z[2] += 1
    if r == 1:
        if s == 0:
            if t == 0:
                y[4] += 1
                z[1] += 1
            if t == 1:
                y[5] += 1
                z[2] += 1
        if s == 1:
            if t == 0:
                y[6] += 1
                z[2] += 1
            if t == 1:
                y[7] += 1
                z[3] += 1
   # print(r, s, t)

print(y)
plt.bar(outcomes,y)
plt.xlabel('outcomes')
plt.ylabel('occurrences')
plt.title('coin flips and stuff')
```
