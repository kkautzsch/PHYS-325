```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize

#Constants
R = 10000 #Ohms
R_L = 33 #Ohms
C = 0.1*10**(-6) #microFarads
L = 10*10**(-3) #milliHenry's

#Functions

#Voltage Function
def V(w):
    return (np.sqrt((1 + (R*R_L*C)/L)**2 + (R*w*C - R/(w*L))**2))**(-1)

#Phase Angle Function
def phi(w):
    return -np.arctan((R*w*C - R/(w*L))/(1 + (R*R_L*C)/L))

#Range for omega
w = np.arange(0.01,60e3,0.01)

#Funtion to find omega value
def f(x): 
    return V(x)
maxOmega = scipy.optimize.fmin(lambda x: -f(x), 0)

print('Value for omega',maxOmega)

plt.plot(w,V(w))
plt.title('$|V_{out}/V_{in}|$ vs. $\omega_{0}$')
plt.xlabel('$\omega (Hz)$')
plt.ylabel('$|V_{out}/V_{in}|$')
plt.show()
```

```python
plt.plot(w,phi(w))
plt.title('$\Delta\phi$ vs. $\omega_{0}$')
plt.xlabel('$\omega (Hz)$')
plt.ylabel('$\Delta\phi$')
plt.show()
```

```python
#Q-Factor
Q = maxOmega*R*C
print('Q = ',Q)
```
